## Описание проекта
Интернет-магазин _**«В один клик»**_ продаёт разные товары: для детей, для дома, мелкую бытовую технику, косметику и даже продукты. Отчёт магазина за прошлый период показал, что активность покупателей начала снижаться. Привлекать новых клиентов уже не так эффективно: о магазине и так знает большая часть целевой аудитории. Возможный выход — удерживать активность постоянных клиентов. Компания хочет этого достигнуть с помощью анализа персонализированных предложений: нужно разработать решение, которое позволит персонализировать предложения постоянным клиентам, чтобы увеличить их покупательскую активность.

### Решение задачи
Для решения задачи нужно:
1. Нужно промаркировать уровень финансовой активности постоянных покупателей. В компании принято выделять два уровня активности: «снизилась», если клиент стал покупать меньше товаров, и «прежний уровень».


2. Нужно собрать данные по клиентам по следующим группам:
     - Признаки, которые описывают коммуникацию сотрудников компании с клиентом.
     - Признаки, которые описывают продуктовое поведение покупателя. Например, какие товары покупает и как часто.
     - Признаки, которые описывают покупательское поведение клиента. Например, сколько тратил в магазине.
     - Признаки, которые описывают поведение покупателя на сайте. Например, как много страниц просматривает и сколько времени проводит на сайте.


3. Нужно разработать модель, которая предскажет вероятность снижения покупательской активности.

4. Нужно выделить сегмент покупателей, проанализировать его и предложить, как увеличить его покупательскую активность. 

5. Нужно использовать данные моделирования, данные о прибыли покупателей и исходные данные для анализа.

### Описание данных
Данные для работы находятся в нескольких таблицах. 


1. **`market_file.csv`** - таблица, которая содержит данные о поведении покупателя на сайте, о коммуникациях с покупателем и его продуктовом поведении:
    - `id` — номер покупателя в корпоративной базе данных.
    - `Покупательская активность` — рассчитанный класс покупательской активности (целевой признак): «снизилась» или «прежний уровень».
    - `Тип сервиса` — уровень сервиса, например «премиум» и «стандарт».
    - `Разрешить сообщать` — информация о том, можно ли присылать покупателю дополнительные предложения о товаре. Согласие на это даёт покупатель.
    - `Маркет_актив_6_мес` — среднемесячное значение маркетинговых коммуникаций компании, которое приходилось на покупателя за последние 6 месяцев. 
    - `Маркет_актив_тек_мес` — количество маркетинговых коммуникаций в текущем месяце.
    - `Длительность` — значение, которое показывает, сколько дней прошло с момента регистрации покупателя на сайте.
    - `Акционные_покупки` — среднемесячная доля покупок по акции от общего числа покупок за последние 6 месяцев.
    - `Популярная_категория` — самая популярная категория товаров у покупателя за последние 6 месяцев.
    - `Средний_просмотр_категорий_за_визит` — показывает, сколько в среднем категорий покупатель просмотрел за визит в течение последнего месяца.
    - `Неоплаченные_продукты_штук_квартал` — общее число неоплаченных товаров в корзине за последние 3 месяца.
    - `Ошибка_сервиса` — число сбоев, которые коснулись покупателя во время посещения сайта.
    - `Страниц_за_визит` — среднее количество страниц, которые просмотрел покупатель за один визит на сайт за последние 3 месяца.


2. **`market_money.csv`** - таблица с данными о выручке, которую получает магазин с покупателя, то есть сколько покупатель всего потратил за период взаимодействия с сайтом.
    - `id` — номер покупателя в корпоративной базе данных.
    - `Период` — название периода, во время которого зафиксирована выручка. Например, 'текущий_месяц' или 'предыдущий_месяц'.
    - `Выручка` — сумма выручки за период.

3. **`market_time.csv`** - таблица с данными о времени (в минутах), которое покупатель провёл на сайте в течение периода.
    - `id` — номер покупателя в корпоративной базе данных.
    - `Период` — название периода, во время которого зафиксировано общее время.
    - `минут` — значение времени, проведённого на сайте, в минутах.

4. **`money.csv`** - таблица с данными о среднемесячной прибыли покупателя за последние 3 месяца: какую прибыль получает магазин от продаж каждому покупателю.
    - `id` — номер покупателя в корпоративной базе данных.
    - `Прибыль` — значение прибыли.

Теперь нужно проанализировать распределение всех признаков.

```python
# Функция для визуализации гистограмм
def hist_plots(df, column, range_left=(0, 1), range_right=(0, 100), bins=100, left_right=True):
    print('Среднее значение - ', df[column].mean())
    print('Минимум и Максимум - ', (df[column].min(), df[column].max()))
    
    ax = df[column].hist(bins=bins, figsize=(15, 4))
    plt.title(f'Гистограммы признака "{column}"')
    plt.xlabel(column)
    plt.ylabel('Частота')
    plt.show()
    
    if left_right:
        ax = df[column].hist(bins=bins, figsize=(15, 4), range=range_left)
        plt.xlabel(column)
        plt.ylabel('Частота')
        plt.show()
    
        ax = df[column].hist(bins=bins, figsize=(15, 4), range=range_right)
        plt.xlabel(column)
        plt.ylabel('Частота')
        plt.show()

    ax = df[column].plot(kind='box', vert=False, figsize=(15, 4))
    plt.show()
```

## Объединение таблиц

Теперь нужно объединить таблицы. Это можно сделать по столбцу `id` - идентификатору покупателя.

```python
data_market_money.shape
```

```python
data_market_money.head()
```

Данные о выручке и времени на сайте находятся в одном столбце для всех периодов. В итоговой таблице лучше сделать отдельный столбец для каждого периода.

```python
# Добавляем дополнительные столбцы для выручки за каждый период
data_market_money['выручка_препредыдущий_месяц'] = data_market_money\
                .apply(lambda row: row['выручка'] if row['период'] == 'препредыдущий_месяц' else 0, axis=1)

data_market_money['выручка_предыдущий_месяц'] = data_market_money\
                .apply(lambda row: row['выручка'] if row['период'] == 'предыдущий_месяц' else 0, axis=1)

data_market_money['выручка_текущий_месяц'] = data_market_money\
                .apply(lambda row: row['выручка'] if row['период'] == 'текущий_месяц' else 0, axis=1)

data_market_money = data_market_money[['id', 'выручка_препредыдущий_месяц', 'выручка_предыдущий_месяц', 'выручка_текущий_месяц']]\
            .groupby('id')[['выручка_препредыдущий_месяц', 'выручка_предыдущий_месяц', 'выручка_текущий_месяц']].sum()
data_market_money.head()
```

```python
# Добавляем дополнительные столбцы с минутами
data_market_time['минут_предыдуший_месяц'] = data_market_time\
                .apply(lambda row: row['минут'] if row['период'] == 'предыдцщий_месяц' else 0, axis=1)


data_market_time['минут_текущий_месяц'] = data_market_time\
                .apply(lambda row: row['минут'] if row['период'] == 'текущий_месяц' else 0, axis=1)

data_market_time = data_market_time[['id', 'минут_предыдуший_месяц', 'минут_текущий_месяц']]\
            .groupby('id')[['минут_предыдуший_месяц', 'минут_текущий_месяц']].sum()

data_market_time.head(3)
```

Новые столбцы были успешно добавлены.

```python
# Объединяем первые 2 датасета
all_data = data_market_file.merge(data_market_money, on='id', how='inner')

# Присоединяем 3 датасет
all_data = all_data.merge(data_market_time, on='id', how='inner')
```

## Использование пайплайнов

### Создание выборок

```python
X, y = all_data.drop('покупательская_активность', axis=1),  all_data['покупательская_активность']
```

```python
y.value_counts().plot.bar(figsize=(8, 5));
```

Итак, как видно, имеет место дисбаланс классов. Это нужно учесть при формировании выборок и обучении моделей. Разобъём исходные данных на тренировочную, валидационную и тестовую выборки. При этом нужно стратифицировать выборки при разделении. Перед этим закодируем целевой признак.

```python
RANDOM_STATE = 10
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=RANDOM_STATE, stratify=y)

label_enc = LabelEncoder()
y_train = label_enc.fit_transform(y_train)
y_test = label_enc.transform(y_test)
```

```python
print('Размерность тренировочных данных - ', X_train.shape)
#print('Размерность валидационных данных - ', X_val.shape)
print('Размерность тестовых данных - ', X_test.shape)
```

Итак, выборки успешо созданы.

### Создание пайплайнов

```python
cat_columns = ['тип_сервиса', 'разрешить_сообщать', 'популярная_категория',]


num_columns = ['маркет_актив_6_мес', 'маркет_актив_тек_мес', 'длительность', 'акционные_покупки',
               'средний_просмотр_категорий_за_визит', 'неоплаченные_продукты_штук_квартал', 
               'ошибка_сервиса', 'страниц_за_визит', 'выручка_препредыдущий_месяц', 'выручка_предыдущий_месяц',
               'выручка_текущий_месяц', 'минут_предыдуший_месяц', 'минут_текущий_месяц'] 
```

```python
# Пайплайн для обработки данных
preprocessor_pipe = ColumnTransformer([
    ('encoder', OneHotEncoder(drop='first', sparse=False, handle_unknown='error'), cat_columns),
    ('scaler', StandardScaler(), num_columns)   
])

# Пайплайн для обработки + обучения модели
final_pipe = Pipeline([
    ('preprocessor', preprocessor_pipe),
    ('models', KNeighborsClassifier())
])
```

Осуществим подбор моделей, методов скалирования/масштабирования через pipeline'ы.

```python
# Задаём список параметров
params = [
    # 1-ая модель и гиперпараметры
    {'models': [KNeighborsClassifier()], 'models__n_neighbors': range(3, 15), 'models__p': [1, 2],
     'preprocessor__scaler': [StandardScaler(), MinMaxScaler(), RobustScaler()]},

    
    # 2-ая модель и гиперпараметры
    {'models': [DecisionTreeClassifier(random_state=RANDOM_STATE, class_weight='balanced')], 
     'models__max_depth': range(3, 15), 'models__min_samples_split': range(2, 10),
     'models__min_samples_leaf': range(1, 5),
    'preprocessor__encoder': [OrdinalEncoder(handle_unknown="use_encoded_value", unknown_value=np.nan), OneHotEncoder(drop='first', sparse=False, handle_unknown='error')],
    'preprocessor__scaler': [StandardScaler(), MinMaxScaler(), RobustScaler()]},
    
    # 3-ая модель и гиперпараметры
    {'models': [LogisticRegression(random_state=RANDOM_STATE, class_weight="balanced")], 'models__C': [1, 5, 8, 10, 20, 25],
    'preprocessor__scaler': [StandardScaler(), MinMaxScaler(), RobustScaler()]},
 
    
    # 4-ая модель и гиперпараметры
    {'models': [SVC(random_state=RANDOM_STATE, class_weight="balanced")], 'models__C': [1, 5, 8, 10, 20, 25],
     'models__kernel': ['poly', 'rbf', 'sigmoid'], 'models__degree': range(2, 5), 
     'preprocessor__scaler': [StandardScaler(), MinMaxScaler(), RobustScaler()]}
]

```

Нужно решить задачу бинарной классификации при несбалансированных классах. В данном случае лучше использовать метрику **F1-score**. Она снижается, когда модель из-за дисбаланса плохо предсказывает минорный класс. 

```python
random_search = RandomizedSearchCV(final_pipe, param_distributions=params, n_iter=30, scoring='f1', random_state=RANDOM_STATE)
```

```python
random_search.fit(X_train, y_train)
```

```python
random_search.best_score_
```

Метрика **`f1_score`** принимает значения в интервале $[0, 1]$. Как видно, у лучшей модели метрика превышает $0,8$. Это хороший результат.

```python
best_model = random_search.best_estimator_
best_model
```

Лучшей моделью оказалось дерево решений с параметрами `class_weight='balanced'`, `max_depth=13`, `min_samples_split=5`. При этом в качестве scaler'а использовался простой `StandardScaler`. Расчитаем значение на тестовой выборки.

```python
y_pred = best_model.predict(X_test)
print('Метрика на тестовых данных - ', f1_score(y_test, y_pred))
```

На тестовых данных метрика оказалась немного больше, чем на кросс-валидации. Рассмотрим значение основных метрик.

```python
pd.DataFrame([f1_score(y_test, y_pred), roc_auc_score(y_test, best_model.predict_proba(X_test)[:, 1]), 
              accuracy_score(y_test, y_pred), precision_score(y_test, y_pred), recall_score(y_test, y_pred)],
             index=['f1_score', 'roc_auc_score', 'accuracy_score', 'precision_score', 'recall_score'], columns=['metrics'])
```

Все метрики достаточно высоки. Однако, метрика `recall` отстаёт от других. Если основная задача - выявить как можно больше элементов положительного класса, то нужно улучшить эту метрику. В целом, модель предсказывает достаточно хорошо.

Далее проведём анализ данных по сегменту и сформулируем предложения по работе с сегментом для увеличения покупательской активности.

```python
def errors(x):
    if x == 0:
        return 'ошибок_нет'
    
    elif 0 < x <= 4:
        return 'мало_ошибок'
    
    elif 4 < x <= 7:
        return 'много_ошибок'
    
    elif x > 7:
        return 'очень_много_ошибок'

def not_paid_goods(x):
    if x <= 3:
        return 'мало'
    
    elif x <= 6:
        return 'много'
    
    elif x > 6:
        return 'очень_много'


def promotions(x):
    if x <= 0.17:
        return 'мало_акций'
    
    elif x <= 0.30:
        return 'нормально_акций'
    
    elif x <= 0.5:
        return 'много_акций'
    
    elif 0.5 < x:
        return 'очень_много_акций'


def marketing_6_mon(x):
    if x <= 3:
        return 'мало_маркетинга'
        
    elif x <= 5:
        return 'среднее_маркетинга'
    
    else:
        return 'много_маркетинга'
        
def marketing_current_mon(x):
    if x <= 3:
        return 'мало_маркетинга'
        
    elif x <= 5:
        return 'среднее_маркетинга'
    
    else:
        return 'много_маркетинга'
    

def total_minutes(x):
    if x <= 25:
        return 'мало_времени'
    
    elif x <= 35:
        return 'среднее_количество_времени'
    
    else:
        return 'много_времени'



def registered_time(x):
    if x < 360:
        return 'меньше_года'  
    
    elif 360 <= x < 720:
        return 'до_2_лет'
        
    elif x >= 720:
        return 'больше_2_лет'


def pages_per_visit(x):
    if 0 <= x < 5:
        return 'неактивный_пользователь'
    
    elif 5 <= x < 10:
        return 'активный_пользователь'
    
    elif x >= 10:
        return 'очень_активный_пользователь'
```

Создадим группы для более удобного анализа выделенного сегмента.

```python
clients_segment['время_группы'] = clients_segment['совокупные_минуты'].apply(total_minutes)
clients_segment['акционные_покупки_группы'] = clients_segment['акционные_покупки'].apply(promotions)
clients_segment['длительность_группы'] = clients_segment['длительность'].apply(registered_time)
clients_segment['страниц_группы'] = clients_segment['страниц_за_визит'].apply(pages_per_visit)
clients_segment['маркет_актив_6_мес_группы'] = clients_segment['маркет_актив_6_мес'].apply(marketing_6_mon)
clients_segment['маркет_актив_тек_группы'] = clients_segment['маркет_актив_тек_мес'].apply(marketing_current_mon)
clients_segment['неоплаченные_продукты_штук_квартал_группы'] = clients_segment['неоплаченные_продукты_штук_квартал']\
                .apply(not_paid_goods)
clients_segment['ошибка_сервиса_группы'] = clients_segment['ошибка_сервиса'].apply(errors) 
```
